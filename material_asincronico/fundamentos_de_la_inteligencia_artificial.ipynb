# Fundamentos de la Inteligencia Artificial

## Introducción a la Inteligencia Artificial

La IA es la capacidad de una máquina para aprender patrones y predicciones.

**Tipos de IA**

* Estrecha: dedicado a la máquina, esta toma decisiones por si sola (de una sola tarea).
* Amplia: mezcla el trabajo de máquina con supervisión humana.
* General: hacer todo sin supervisión, igualando capacidades humanas (no existe aún).

**Proceso de aprendizaje automático**

* No requiere de datos estructurados
* Solo requiere de los caminos del árbol los cuales podrá "pronosticar"

**Sistemas**

* Sistema determinista: requiere de una gran base de datos estructurada con las rutas. Para avanzar, ve si la ruta sirve o no (SI/NO).
* Sistema probabilístico: no usa SI/NO, mas bien contruye todas las posibles rutas hacia el destino y las compara.

**Tipos de Aprendizaje**

* Aprendizaje Supervisado: a partir de ejemplos, puede usar eiquetas para que aprenda a reconocer.
* Aprendizaje no Supervisado: se le entrega toda la información para que aprenda sola. Realiza distintas posibles combinaciones y no usa etiquetas predeterminadas (crea etiquetas por si misma).
* Aprendizaje de Refuerzo: prueba y error, se le entrega los datos y se le "recompensa" al encontrar una solución.

## Procesamiento del lenguaje natural y visión por ordenador

**Pasos de un debate**

**1. Aprender y comprender el tema:** ingerir varios miles de millones de párrafos de libros (corpus). Luego estructurar el contenido para relacionar conceptos.

**2. Crear una posición:** desarrollar discurso compuesto por breves fragmentos de texto a partir del corpus. Debe detallar su posición sobre el tema.

**3. Organizar sus pruebas:** conocer el significado profundo de las relaciones con el tema. Decidir pruebas sólidas y organizarlas por tema.

**4. Responder a tu oponente:** escuchar argumentos y refutarlos, consolidando argumentos.

Para entender el lenguaje humano se necesita procesamiento del lenguaje natual (NLP).

**Procesamiento de lenguaje natural**

* El lenguaje humano no está estructurado.
* Cuesta manejar datos no estrcuturados.
* Para ello se usa segmentación de oraciones, fragmentando de información denominados señales (tokens), que se pueden clasificar. Se peuden clasificar entidades, siempre y cuando la segmentación sea sustantivo.
* Otra son las relaciones, son un grupo de dos o más entidades que tienen una conexión entre sí. 
* Una vez clasificadas entidades y relaciones, estructurar la información.
* Conceptos es algo implícito en una oración. Inplica emparejar ideas en lugar de palabras.

**Detección de emociones y análisis de opinión**

Sirve para entender si el lenguaje expera ira, felicidad, tristeza. Por ejemplo, en una encuesta de satisfacción.

**Problema de la clasificación**

* Lenguaje humano tiene términos ambiguos o con doble sentido. 
* La IA identifica la frecuencia en que se vinculan las oraciones. El sistema mejora poco a poco en la clasificación, pero puede cometer errores (en disminución).

**Estructura de un chatbot**

* Los chatbots trabajan con pocos datos. 
* Su escala es limitada.
* Tiene frontend y backend
* Backend/Algoritmos clasificadores: Pueden correlacionar muchas formas distintas de formular una pregunta con un conjunto muy reducido de respuestas. Por ejemplo, ante muchas preguntas parecidas puede haber una sola respuesta.

**Intenciones, entidades y diálogo**

Los chatbots entienden una pregunta dividiéndola en parttes y relacionando esas partes con cosas de su memoria. El objetivo es identificar entidades e intenciones, y utilizar lo que se ha encontrado para activar un diálogo.

* Intención: la razón por la cual un usuario contacta al chatbot. Primero hablar con alguien que haya respondido a muchas preguntas por teléfono, haciendo una lista de todas las formas que una persona puede preguntar del tema.

Una vez se tenga la lista de intenciones que debe cumplir el chatbot, se continúa.

* Entidad: es un sustantivo. Es una persona, lugar u objeto.

Si un usuario pregunta: “¿Cuál es el horario del restaurante de Austin?”, la intención es proporcionar el horario de apertura y Austin es la entidad. Un chatbot necesita una lista completa de entidades para resultar útil.

* Diálogo: es un diagrama de flujo, una estructura de árbol IF / THEN que muestra cómo responderá una máquina a las intenciones del usuario. Representa cada una de las posibles palabras o frases que puede introducir un usuario, la respuesta adecuada para el chatbot y las posibles respuestas posteriores.

**La IA clasifica las imágenes**

**Redes neuronales convolucionales**

Las fotos y videos crean millones de pixeles, expresados en números. Lo complicado es analizar estos datos. Los pixeles son cifran grandes que abruman a los sistemas de inteligencia artificial. Para evitar desbordar un sistema, se utiliza red neuronal convolucional (CNN), permite identificar objetos en una imagen, como en el reconocimiento facial.

**Un sistema de reconocimiento visual puede utilizar una red generativa adversarial (GAN) para crear nuevos dibujos y fotos**

Para generar imágenes, se entrentan dos CNN, esto se denomina red generativa artificial GAN. Las CNN luchan entre sí hasta que una de ellas genera la imagen.
