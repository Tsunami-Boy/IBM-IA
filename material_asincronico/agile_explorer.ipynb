# Agile Explorer

**Manifiesto**
Interacciones antes que procesos, herramientas en funcionamiento antes que foducmentación detallada.

**Valores**
* Confianza: capacidad de ser vulnerable con los demás, como cometer errores, pedir ayuda o asumir riesgos.
* Empatía: comprender a los demás.
* Respeto: mostrar consideración y estima por sobre los demás.
* Transparencia: tener la mente abierta, escuchar nuevas ideas.
* Valentía: perseverar y esforzarse.

**Pasos**
1) Definir el Problema: definir evita errores y sesgos. Objetivos.
2) Determinar el Resultado para el Usuario: el resultado o valor ofrecido al usuario final. Criterios de éxito.
3) Mantener a la Vista el Resultado al Usuario: debe orientar y confirmar resultados definidos. Garantiza valor real.

**Prácticas Centrales de Agile**
1) Contrato social: como colaborarán, se comunicarán y manejarán desacuerdos.
2) Canicas de estado de ánimo: evaluar el estado de ánimo general cada día.
3) Muro de Trabajo: muestra tareas pendientes y complejidades. Todos informados.
4) Reuniones breves de progreso: cada día se prepara reunión, se comparte que se logró ayer, que se hará hoy y los obstáculos.
5) Retrospetiva: semanalmente se analiza que se hizo bien, que se hizo mal y cómo mejorar el enfoque.
6) Muestra: se hace una muestra al arquitecto, dando retrospectiva y recomendaciones de diseño. Ayuda a la siguiente iteración.
7) Shuhari: comienza con ideas, experimenta nuevas ideas y las mejoras. Este proceso lo vuelve experto.

**Contrato Social**
Se recomienda generar un tableco con adhesivo, donde los miembros del equipo escriben qué es lo que desean del equipo, sus proyecciones y qué esperan.

**Canicas del estado de ánimo**
A partir de canicas verdes y rojas, diariamente al final del día los miembros del equipo insertan una canica dependiendo de su estado de ánimo luego del día laboral. Al día siguiente se observa si el equipo se sintió agusto o incómodo, y se habla al respecto.

**Muro de trabajo**
Similar a Jira, contextualiza al equipo respecto a las tareas por hacer, quien está asignado, las que están en curso y las finalizadas.

**Reuniones breves de progreso**
Se realizan reuniones diarias, abarcando temas como qué se hizo, qué falta por hgacer e impedimentos. Se hace al empezar el día.

**Retrospectiva**
Reunión que abarca lo que salió bien, lo que salió mal. Básicamente retroalimentación de lo realizado.

**Patrones Ágiles**
Brindan solución a los desafíos comunes que enfrentan los equipos.

* Patrón de gestión: dirección que seguirá la empresa.
* Patrón de operaciones: trabajo diario continuo que implica procesos únicos y repetibles.
* Patrón de programa: trabajos de proyecto que tienen inicio y final.
