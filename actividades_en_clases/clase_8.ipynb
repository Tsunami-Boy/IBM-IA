# Visualización

Transforma datos en representaciones gráficas interpretables.


* Revela patrones no detectables
* Identifica valores atípicos y distribuciones anómalas
* Comunica hallazgos de forma comprensible
* Valida hipótesis sobre comportamiento de datos

## Matplotlib

```python
# pip install matplotlib
import matplotlib.pyplot as plt
```

## Elementos Fundamentales

```python
# Crear Figura
plt.figure(figsize=(10, 6))

# Agregar ejes
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]
plt.plot(x,y, marker='o', color='b', label='Datos de ejemplo')
plt.legend() # Si se usa label en plot, agregar leyenda

# Títulos y etiquetas
plt.title('Gráfico de Ejemplo')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Mostrar gráfico
plt.show()
```

## Gráfico de línea

```python
# Crear Figura
plt.figure(figsize=(10, 6))

# Agregar ejes
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]
# marker: forma del punto, color: color del punto/linea, label: etiqueta para la leyenda
plt.plot(x,y, marker='o', color='b', label='Datos de ejemplo')
plt.legend() # Si se usa label en plot, agregar leyenda

```

## Gráfico de barras

```python
# Crear Figura
plt.figure(figsize=(5, 3))

# Vertical
cat = ["A", "B", "C", "D"]
val1 = [5, 7, 3, 4]
val2 = [6, 9, 2, 5]
plt.bar(cat, val1)
```

```python
# Horizontal
plt.figure(figsize=(5, 3))
plt.barh(cat, val2, color='orange')
```

```python
# Agrupadas
plt.figure(figsize=(5, 3))
plt.bar(cat, val1)
plt.bar(cat, val2)  # Apiladas
```

## Gráfico de Dispersión

```python
# Crear Figura
plt.figure(figsize=(5, 3))

# Datos
cat = ["A", "B", "C", "D"]
val1 = [5, 7, 3, 4]
val2 = [6, 9, 2, 5]

# Gráfico de dispersión
plt.scatter(cat, val1)
```

```python
# Tamaño variable
sizes = [50, 100, 200, 300]
plt.scatter(cat, val2, s=sizes)
```

```python
# Colores por grupo
colors = ['red', 'blue', 'green', 'orange']
plt.scatter(cat, val1, s=100, c=colors)
```

## Histogramas

```python
# Crear Figura
plt.figure(figsize=(5, 3))

# Datos
cat = ["A", "B", "C", "D"]
val1 = [5, 7, 3, 4]
val2 = [6, 9, 2, 5]

# Histogramas
# alpha: transparencia, label: etiqueta para la leyenda
plt.hist(val1, alpha=0.5, label='Val1')
```

```python
# Histograma con intervalos
plt.hist(val2, bins=3, alpha=0.8, label='Val2')
plt.legend()
```

## Buenas prácticas

* Títulos: claros y con contexto
* Colores: paletas consistentes y accesibles
* Texto: legible y con buen contraste
* Ejes: escalas completas y proporcionales
* Claridad: gráficos simples y directos
