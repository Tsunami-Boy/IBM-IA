**Registro de Temperaturas**

1. Solicitar la temperatura de 5 días consecutivos.
2. Mostrar la temperatura promedio, máxima y mínimo de la semana.
3. Contar cuántos días tuvieron temperatura alta (mayor a 25°C).
4. Mostrar un resumen completo de los datos registrados.

```python
lista = []
dia = 1
while (len(lista) < 5):
    lista.append(int(input(f"Digite la temperatura del día {dia}:")))
    dia+=1

promedio = sum(lista)/len(lista)
temp_max = max(lista)
temp_min = min(lista)

print(f"El promedio de las temperaturas es {promedio}°C")
print(f"La temperatura máxima es {temp_max}°C")
print(f"La temperatura mínima es {temp_min}°C")

temp_may_25 = 0
for i in lista:
    if(i > 25):
        temp_may_25 += 1

print(f"La cantidad de días con temperatura mayor a 25°C es {temp_may_25}")
```
