## Distribuciones de datos

Muestran la forma en que se organizan los valores dentro de un conjunto de datos

* Normal: campana simétrica
* Sesgada: cola hacia un lado (escalonada)
* Bimodal: dos picos de frecuencia (en los extremos)
* Multimodal: múltiples picos de frecuencia
* Uniforme: frecuencias similares

## Identificación de distribuciín

Se deduce comaprando media y mediana

* Moda: el número que más se repite en un conjunto de datos
* Media: promedio de un conjunto de datos (depende de los valores de los datos)
* Mediana: el valor central en un conjunto de datos (depende de la cantidad de datos ordenados)

Tipos de distribución:

* Normal: Mediana = Media (alturas, pesos)
* Sesgada: Mediana muy diferente que Media (ingresos, precios)
* Bimodal: Depende de los picos (horarios de tráfico)
* Multimodal: Variable según picos (preferencias múltiples)
* Uniforme: Media = Mediana (números aleatorios)

Tener en consideración:

* Rango: diferencia entre el más alto y el más pequeño
* Varianza: suma de los cuadrados de las distancias entre la media y cada elemento de la población, dividido en el número de elementos (valor alto indica mayor dispersión).

## Correlaciones

Medida de qué tanto dos variables cambian juntas

* Valores entre -1 (inverso) y +1 (directo)
* 0 indica que no hay relación lineal

```python
# Correlaciones

## Limpiar primero
df['CO2 (kt)'] = pd.to_numeric(
	df['CO2 (kt)'].astype(str)
	  .str.replace('.', '', regex=False)   # quitar separador de miles
	  .str.replace(',', '.', regex=False), # convertir coma decimal a punto
	errors='coerce'                        # convertir valores no convertibles a NaN
)

df['CO2 per cápita (toneladas métricas)'] = pd.to_numeric(
	df['CO2 per cápita (toneladas métricas)'].astype(str)
	  .str.replace('.', '', regex=False)   # quitar separador de miles
	  .str.replace(',', '.', regex=False), # convertir coma decimal a punto
	errors='coerce'                        # convertir valores no convertibles a NaN
)


df[['CO2 (kt)', 'CO2 per cápita (toneladas métricas)']].corr()
```

## Evaluación de confiabilidad

**Desviación estandar:**
* Baja: datos consistentes, resultados predecibles.
* Alta: datos dispersos, mayor incertidumbre.

**Outliers:**
* Analizar antes de eliminarlos.
