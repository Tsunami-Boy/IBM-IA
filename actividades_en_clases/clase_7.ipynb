```python
# Estadísticas agrupadas
'''
En este caso, agrupa todas las regiones y calcula la media y la suma del año para cada región.
'''
df.groupby('Región').agg({'Año': ['mean', 'sum']})
```

**Interferencia estadística:** emplear datos de una muestra para hacer estimaciones y probar hipótesis acerca de las características de una población.

**Población:** cuando se examina un grupo entero o universo completo de observaciones

**Muestra:** cuando se examina una pequeña parte del grupo

**Distribución de frecuencias:**
* Forma de presentación de los datos, debe facilitar su tratamiento y permitir diversas comprensiones.
* Tabla de datos con base en observaciones.
* Número de casos que pertenecen a un valor determinado.

**Histogramas:**
Gráfico de distribución de frecuencias, que se contruye con rectángulos de superficie proporcional al producto de la amplitud por la frecuencia absoluta de cada uno de los intervalos de clase (gráfico de barras).

**Tendencia Central:** punto medio de una distribución. OJO: el sesgo se produce cuando al trazar una linea vertical que pase por el punto más alto de la curva, dividirá su área en dos partes que no son iguales.


## Dsitribuciones de datos

Muestran la forma en que se organizan los valores dentro de un conjunto de datos

* Normal: campana simétrica
* Sesgada_ cola hacia un lado (escalonada)
* Bimodal: dos picos de frecuencia (en los extremos)
* Multimodal: múltiples picos de frecuencia
* Uniforme: frecuencias similares
