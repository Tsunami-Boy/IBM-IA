¿Cuántos años tienes?

Crea un programa interactivo que calcule la edad basándose en el año de nacimiento.
1. Solicita al usuario su nombre y año de nacimiento
2. Calcula la edad basándose en el año actual y muestra qué tipo de dato es
3. Convierte la entrada del usuario para poder hacer cálculos
4. Determina si la persona es mayor de edad y guarda esa información
5. Muestra un saludo personalizado con el nombre y la edad calculada

```python
nombre = input("¿Cuál es tu nombre? ")
anio = int(input(f"Hola, {nombre}! ¿En qué año naciste? "))
edad = 2025 - anio
print(type(edad))
if(edad < 18):
    print(f"{nombre} tienes {edad} años, eres menor de edad.")
else:
    print(f"{nombre} tienes {edad} años, eres mayor de edad.")
```

Lista de compras

Crea un programa interactivo que registre productos y calcule información de
la compra.
1. Pedir al usuario el nombre y precio de 3 productos
2. Guardar los datos en una estructura que permita acceder tanto al
nombre como al precio
3. Calcular el total a pagar y determinar el producto más caro
4. Mostrar la lista de productos, el total y cuál fue el producto de mayor
costo

```python
print("Bienvenid@ al programa de productos!!!")
name_1 = input("Ingrese el nombre del producto 1: ")
price_1 = int(input("Ingrese el precio del producto 1: "))
name_2 = input("Ingrese el nombre del producto 2: ")
price_2 = int(input("Ingrese el precio del producto 2: "))
name_3 = input("Ingrese el nombre del producto 3: ")
price_3 = int(input("Ingrese el precio del producto 3: "))

diccionario_productos = {}
diccionario_productos[name_1] = price_1
diccionario_productos[name_2] = price_2
diccionario_productos[name_3] = price_3

total = price_1 + price_2 + price_3

producto_caro = max(price_1, price_2, price_3)

print(f"El total de la compra es: {total}")
print(f"El producto más caro cuesta: {producto_caro}")
print(f"La lista de productos es la siguiente: {diccionario_productos}")
```
