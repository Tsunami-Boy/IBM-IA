# Rendimiento E-Commerce

* Identificar mes con mayor eficiencia (ventas/gasto publicidad)
* Determinar mes con mejor tasa de conversión y analizar causa
* Calcular ticket promedio (ventas/productos) por mes
* Evaluar relación entre visitantes y ventas

```python
import pandas as pd
df = pd.read_csv('E-commerce.csv', encoding="latin1", sep="\t")
df
```

```python
df['Visitantes'] = pd.to_numeric(df['Visitantes'].astype(str)
	                            .str.replace('.', '', regex=False)  # quitar separador de miles
	                            .str.replace(',', '.', regex=False), # convertir coma decimal a punto        
                                errors='coerce')
df['Gasto Publicidad ($)'] = pd.to_numeric(df['Gasto Publicidad ($)'].astype(str)
	                            .str.replace('.', '', regex=False)  # quitar separador de miles
	                            .str.replace(',', '.', regex=False), # convertir coma decimal a punto        
                                errors='coerce')
df['Ventas ($)'] = pd.to_numeric(df['Ventas ($)'].astype(str)
	                            .str.replace('.', '', regex=False)  # quitar separador de miles
	                            .str.replace(',', '.', regex=False), # convertir coma decimal a punto        
                                errors='coerce')
df['Conversión (%)'] = pd.to_numeric(df['Conversión (%)'],       
                                errors='coerce') 
df.info()
```

```python
# Mes con mayor eficiencia (ventas/gasto publicidad)
df['Eficiencia'] = df['Ventas ($)'] / df['Gasto Publicidad ($)']
mes_mayor_eficiencia = df.loc[df['Eficiencia'].idxmax(), 'Mes']
# df.loc muestra la columa (Mes) segun el id (idmax de Eficiencia)
mes_mayor_eficiencia
```

```python
# Mes con mejor tasa de conversión
mejor_tasa = df.loc[df['Conversión (%)'].idxmin(), 'Conversión (%)']
mejor_tasa
```
