Acabas de abrir tu primera tienda y quieres analizar tus primeras 10
ventas para entender el arranque del negocio.
1. Solicitar al usuario el monto de las primeras 10 ventas
2. Calcular el promedio de estas ventas iniciales
3. Identificar cuáles ventas estuvieron por encima del promedio
4. Calcular el total recaudado en estas primeras ventas
5. Determinar cuál fue tu mejor y peor venta inicial

```python
import numpy as np

ventas = []
while len(ventas) < 10:
    inp = input(f"Ingrese la venta del día {len(ventas)+1}: ")
    ventas.append(int(inp))

array_ventas = np.array(ventas)

promedio = np.mean(array_ventas)

array_booleano = array_ventas > promedio # lista de 1s y 0s
ventas_mayor_al_promedio = np.sum(array_booleano)

total = np.sum(array_ventas)

mejor_venta = np.max(array_ventas)
peor_venta = np.min(array_ventas)

print(f"El promedio de ventas es: {promedio}")
print(f"La cantidad de días con ventas mayores al promedio es: {ventas_mayor_al_promedio}")
print(f"El total de ventas es: {total}")
print(f"La mejor venta fue: {mejor_venta}")
print(f"La peor venta fue: {peor_venta}")
```
