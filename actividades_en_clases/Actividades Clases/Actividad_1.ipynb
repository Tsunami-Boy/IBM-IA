Lista de compras

Crea un programa interactivo que registre productos y calcule información de
la compra.
1. Pedir al usuario el nombre y precio de 3 productos
2. Guardar los datos en una estructura que permita acceder tanto al
nombre como al precio
3. Calcular el total a pagar y determinar el producto más caro
4. Mostrar la lista de productos, el total y cuál fue el producto de mayor
costo

```python
print("Bienvenid@ al programa de productos!")
name_product_1 = input("Ingrese el nombre del producto 1: ")
price_product_1 = int(input(f"Ingrese el precio de {name_product_1}: \n"))

name_product_2 = input("Ingrese el nombre del producto 2: ")
price_product_2 = int(input(f"Ingrese el precio de {name_product_2}: \n"))

name_product_3 = input("Ingrese el nombre del producto 3: ")
price_product_3 = int(input(f"Ingrese el precio de {name_product_3}: \n"))

products = {}
products[name_product_1] = price_product_1
products[name_product_2] = price_product_2
products[name_product_3] = price_product_3

print(f"Los productos ingresados son: {products}")

total = price_product_1 + price_product_2 + price_product_3

max_product_1 = price_product_1 > price_product_2 and price_product_1 > price_product_3
max_product_2 = price_product_2 > price_product_1 and price_product_2 > price_product_3
max_product_3 = price_product_3 > price_product_1 and price_product_3 > price_product_2

print(f"El total de la compra es: {total}")

print(f"¿{name_product_1} es el producto más caro? -> {max_product_1}")
print(f"¿{name_product_2} es el producto más caro? -> {max_product_2}")
print(f"¿{name_product_3} es el producto más caro? -> {max_product_3}")


```
