# Aspectos importantes para tomar en cuenta

## Reglas de nomenclatura

* Palabras separadas por guiones bajos
* Sesible a mayúsculas
* No iniciar con números

## Funciones

```python
# Redondeo
redondeo = round(3.14159, 2)
print(redondeo)  # Salida: 3.14
```

```python
# Interpolación
nombre = "Ana"
print(f"Hola, {nombre}!")  # Salida: Hola, Ana!
```

```python
# Eliminar espacios
texto = "   Hola, Mundo!   "
texto_limpio = texto.strip()
print(f"'{texto_limpio}'")  # Salida: 'Hola, Mundo!'
```

```python
# Valor máximo y mínimo
numeros = [3, 1, 4, 1, 5, 9]
maximo = max(numeros)
minimo = min(numeros)
print(f"Máximo: {maximo}, Mínimo: {minimo}")  # Salida: Máximo: 9, Mínimo: 1
```
