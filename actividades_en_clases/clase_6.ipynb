## Tratamiento de valores faltantes

```python
# Identificar de valores faltantes
df_csv.isnull().sum()
df_csv.isna().sum()
```

```python
# Elimina filas con valores faltantes
df_csv.dropna()
```

```python
# Rellena valores faltantes con 0
df_csv.fillna(0)
```

```python
# Rellena valores faltantes con la mediana de la columna
columna = 'Año'
df_csv.fillna(df_csv[columna].median())
```

## Eliminación de duplicados

```python
# Marca de duplicados
df_csv.duplicated()
```

```python
# Eliminación de duplicados
df_csv.drop_duplicates()
```

```python
# Eliminación de duplicados basados en columnas específicas
columnas = ['Región', 'Año']
df_csv.drop_duplicates(subset=columnas)
```

## Inconsistencias de formato

```python
# Inconsistencias de formato
# str.lower() # Convierte a minúsculas
# str.upper() # Convierte a mayúsculas
# str.strip() # Elimina espacios en blanco al inicio y al final
# str.replace(' ', '_') # Reemplaza espacios en blanco por guiones bajos

# Aplicar formato a una columna específica|
columna_texto = 'Región'
df_csv[columna_texto] = df_csv[columna_texto].str.lower().str.strip()
```

```python
# Convertir una columna a formato de fecha
# pd.to_datetime() 
'''
Aplicar formato de fecha a una columna específica (en esta base no hay fechas)
columna_fecha = 'Fecha'
df_csv[columna_fecha] = pd.to_datetime(df_csv[columna_fecha], format='%Y-%m-%d')
'''
```

```python
# Ajustar separadores decimales
# pd.to_numeric()
'''
Aplicar formato numérico a una columna específica (en esta base no hay columnas con separadores decimales)
columna_numerica = 'Valor'
df_csv[columna_numerica] = pd.to_numeric(df_csv[columna_numerica].str.replace(',', '.'))
'''
```
